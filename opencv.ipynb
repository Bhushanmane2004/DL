```python
import cv2
from google.colab.patches import cv2_imshow  # Importing cv2_imshow for Colab

# Path to the Haar cascade file
cascade_path = "/content/haarcascade_frontalface_default (1).xml"
classifier = cv2.CascadeClassifier(cascade_path)

# Path to the image file
image_path = "/content/WIN_20230320_14_24_49_Pro.jpg"

# Load the image
image = cv2.imread(image_path)

if image is None:
    print("Could not load the image.")
    exit()

# Convert the image to RGB (if needed)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Detect faces
bboxes = classifier.detectMultiScale(image_rgb)

# Draw rectangles around detected faces
for (x, y, w, h) in bboxes:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Display the result using cv2_imshow
cv2_imshow(image)
```
